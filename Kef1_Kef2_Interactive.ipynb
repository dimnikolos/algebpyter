# Κεφάλαιο 2: Φυσικοί Αριθμοί και Πράξεις

## Οι αριθμοί και η Python

Οι φυσικοί αριθμοί είναι οι αριθμοί ($0, 1, 2, 3, 4, \ldots, 100, \ldots, 1000, \ldots$).

Η Python καταλαβαίνει τους αριθμούς, αλλά **απεχθάνεται τις τελείες**! Αν θες να γράψεις το ένα εκατομμύριο, ενώ στο χαρτί γράφεις $1.000.000$, στην Python πρέπει να το γράψεις ενωμένο: `1000000`.

## Πρόσθεση, αφαίρεση και πολλαπλασιασμός (Σελίδα 16)

> **ΑΣΚΗΣΗ 1:** Να υπολογιστούν τα γινόμενα:
> α) $35 \cdot 10$
> β) $421 \cdot 100$
> γ) $5 \cdot 1.000$

Ο πολλαπλασιασμός στην Python γίνεται με το αστεράκι `*`. Ας βάλουμε την Python να κάνει τις πράξεις για εμάς!

```python
print("α) Το αποτέλεσμα είναι:", 35 * 10)
print("β) Το αποτέλεσμα είναι:", 421 * 100)
print("γ) Το αποτέλεσμα είναι:", 5 * 1000)
```

> **ΑΣΚΗΣΗ 2:** Να εκτελεστούν οι ακόλουθες πράξεις:
> α) $89 \cdot 7 + 89 \cdot 3$
> β) $76 \cdot 13 - 76 \cdot 3$

Όπως και στα Μαθηματικά, έτσι και στην Python, **ο πολλαπλασιασμός προηγείται** της πρόσθεσης και της αφαίρεσης! Η Python ξέρει την προτεραιότητα πράξεων.

```python
print("α)", 89 * 7 + 89 * 3)
print("β)", 76 * 13 - 76 * 3)
```

### Παράδειγμα με παρενθέσεις

Τι γίνεται αν έχουμε παρενθέσεις; Η Python, όπως και εμείς, κάνει πρώτα ό,τι υπάρχει μέσα στην παρένθεση. 

Για παράδειγμα: $60 - (18 - 2)$

```python
print(60 - (18 - 2))
```

## Ένα πραγματικό πρόβλημα (Σελίδα 18)

> **ΠΡΟΒΛΗΜΑ:** Σε ένα αρτοποιείο έφτιαξαν μία μέρα 120 κιλά άσπρο ψωμί, 135 κιλά χωριάτικο και 25 κιλά σικάλεως. 
> Πουλήθηκαν 107 κιλά άσπρο, 112 χωριάτικο και 19 σικάλεως. Πόσα κιλά ψωμί έμειναν απούλητα;

Αντί να τα υπολογίσουμε όλα με το μυαλό μας, μπορούμε να πούμε στην Python να υπολογίσει τη διαφορά (φτιαγμένα μείον πουλημένα) για κάθε είδος και να τα προσθέσει!

```python
aspro_emeine = 120 - 107
xwriatiko_emeine = 135 - 112
sikaleos_emeine = 25 - 19

synolo = aspro_emeine + xwriatiko_emeine + sikaleos_emeine

print("Έμειναν απούλητα:", synolo, "κιλά ψωμί.")
```

*(Σημείωση: Στον παραπάνω κώδικα δώσαμε "ονόματα" -μεταβλητές- σε κάθε αποτέλεσμα για να είναι πιο καθαρή η σκέψη μας!)*

## Δυνάμεις Φυσικών Αριθμών

Στα Μαθηματικά γράφουμε $10^2$ (δέκα στο τετράγωνο). 
Στην Python, η δύναμη συμβολίζεται με **δύο αστεράκια** `**`.

Άρα το $10^2$ γράφεται ως `10**2`.

```python
print("10 εις την δευτέρα (στο τετράγωνο):", 10**2)
print("10 εις την τρίτη (στον κύβο):", 10**3)
print("10 εις την έκτη (ένα εκατομμύριο!):", 10**6)
```

> **ΑΣΚΗΣΗ 3:** Βρες τις τιμές των παραστάσεων: 
> (α) $(6+5)^2$   και   (β) $6^2+5^2$
>
> *Τι παρατηρείς; Είναι το ίδιο;* Ας ρωτήσουμε τον υπολογιστή:

```python
print("Το (6+5)^2 μας κάνει:", (6+5)**2)
print("Ενώ το 6^2 + 5^2 μας κάνει:", 6**2 + 5**2)
```

## Η Υπερδύναμη του Υπολογιστή (Οι Λούπες)

Μέχρι τώρα κάναμε πράξεις που ίσως μπορούσες να κάνεις και στο χαρτί (ή με ένα κομπιουτεράκι). Τι γίνεται όμως αν η άσκηση (Σελίδα 22) ζητάει:

> **ΑΣΚΗΣΗ 4:** Βρες τα τετράγωνα ΟΛΩΝ των αριθμών από το 8 μέχρι το 25.

Στο τετράδιο αυτό θα έπαιρνε πολλή ώρα. Στην Python μπορούμε να της πούμε "Για (for) κάθε αριθμό από το 8 μέχρι το 25, υπολόγισε και τύπωσε το τετράγωνό του".

> *Σημείωση: Στην Python μετράμε μέχρι τον αριθμό που θέλουμε ΣΥΝ ΕΝΑ. Άρα αν θέλουμε μέχρι το 25, θα πούμε `range(8, 26)`.*

```python
for a in range(8, 26):
    print(a**2, end=" ")  # Το end=" " απλά τα τυπώνει δίπλα-δίπλα αντί για μια κάθετη λίστα!
```

Απίστευτο; Σε λιγότερο από ένα δευτερόλεπτο υπολόγισε 18 τετράγωνα.
